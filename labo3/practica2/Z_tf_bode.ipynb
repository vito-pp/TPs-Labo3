```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
# H(z) = (b0 + b1*z^-1 + b2*z^-2) / (1 + a1*z^-1 + a2*z^-2)
b = [0.11574]  # Numerator coefficients
a = [1.0, -0.580337, 0.87824]  # Denominator coefficients
```

```python
w, h = signal.freqz(b, a, worN=8000)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(w / np.pi, 20 * np.log10(abs(h)))
plt.title('Magnitude Response')
plt.xlabel('Normalized Frequency (×π rad/sample)')
plt.ylabel('Magnitude (dB)')
plt.grid()

plt.subplot(1, 2, 2)
plt.plot(w / np.pi, np.angle(h))
plt.title('Phase Response')
plt.xlabel('Normalized Frequency (×π rad/sample)')
plt.ylabel('Phase (radians)')
plt.grid()

plt.tight_layout()
plt.show()
```

```python
zeros, poles, _ = signal.tf2zpk(b, a)

plt.figure(figsize=(6, 6))
plt.plot(0, 0, 'go', label='Zeros')
plt.plot(np.real(poles), np.imag(poles), 'rx', label='Poles')

# Draw unit circle
theta = np.linspace(0, 2 * np.pi, 100)
plt.plot(np.cos(theta), np.sin(theta), 'k--', alpha=0.5)

plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)

plt.title('Pole-Zero Plot in Z-Plane')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.legend()
plt.grid()
plt.axis('equal')
plt.show()
```
